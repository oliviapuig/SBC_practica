```python
import matplotlib.pyplot as plt
import pandas as pd

# Suposem que tens un DataFrame amb la columna "temes_especifics"
# Reemplaça 'nom_del_fitxer.csv' pel nom real del teu fitxer CSV o utilitza una altra font de dades.
df = pd.read_csv('books_clean.csv')

# Compte les freqüències de cada categoria
counts = df['temes_especifics'].value_counts()

# Crea el barplot
plt.bar(counts.index, counts.values)

# Etiquetes i títols
plt.xlabel('Categoria')
plt.ylabel('Freqüència')
plt.title('Barplot de temes_especifics')

plt.xticks(rotation=45, ha='right')
# Mostra el gràfic
plt.show()
```
